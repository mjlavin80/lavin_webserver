```python
words = requests.get('http://ir.dcs.gla.ac.uk/resources/linguistic_utils/stop_words')
stoplist1 = words.text.split("\r\n")

from nltk.corpus import stopwords
stoplist2 = set(stopwords.words('english'))

stoplist1.extend(stoplist2)

fullstops = list(set(stoplist1))

def remove_stops(stoplist, wordlist):
    result = []
    for i in wordlist:
        if i not in stoplist:
                result.append(i)
    return result

def spellcheck(wordlist):
    result = []
    d = enchant.Dict("en_US")
    for i in wordlist:
        if d.check(i) or d.check(i.capitalize()):
            result.append(i)
         
    return result

def clean_text(list_of_texts):
    fully_cleaned =[]
    #normalize ocr errors
    for i in list_of_texts:
        #lowercase all
        ocr_lower = i.lower()
        #tokenize, remove punctuation and numbers, remove tabs, newlines, etc.
        ocr_cleaner = ocr_lower.replace("\n", " ").replace("\t", " ")
        
        ocr_tokens = ocr_cleaner.split(" ")
        
        no_numbers_or_punct = []
        for token in ocr_tokens:
            if token.isalpha():
                no_numbers_or_punct.append(token)
            else:
                
                new_token = ""
                for letter in token:
                    if letter.isalpha():
                        new_token += letter
                if new_token != "":
                    no_numbers_or_punct.append(new_token)  
        
        #almost_ready_before_spellcheck = remove_stops(fullstops, no_numbers_or_punct)
        almost_ready = spellcheck(no_numbers_or_punct)
        #from nltk.stem import WordNetLemmatizer
        #do pos tag before lemmatize
        #lemmatization
        #wordnet_lemmatizer = WordNetLemmatizer()
        #lemmas = []
        #for token in almost_ready:
        #    lemma = wordnet_lemmatizer.lemmatize(token)
        #    lemmas.append(lemma)
        
        #ready = [i for i in lemmas if len(i) > 2]
        #fully_cleaned.append(ready)
        #fully_cleaned.append(no_numbers_or_punct)
        fully_cleaned.append(almost_ready)
    return fully_cleaned
del fullstops[0]

gender_terms = ["mr", "he", "his", "him", "himself", "man", "men", "boy", "boys", "manly", "masculine", "boyish", "father", \
                "brother", "girls", "men", "women", "sisters", "daughters", "brothers", "sons", "wife", "husband", "niece",\
                "uncle", "nephew", "dad", "grandfather", "son", "mrs", "miss", "her", "hers", "she", "herself", "woman",\
                "girl", "nieces", "nephews", \
               "lady", "womanly", "girlish", "girly", "mother", "daughter", "aunt", "niece" "grandmother", "mom", "sister" ]

from nltk.corpus import names 
male = [o.lower() for o in names.words('male.txt')]
female = [o.lower() for o in names.words('female.txt')]

for u in [gender_terms, male, female]:
    for i in u:
        fullstops.append(unicode(i))
    
fullstops = list(set(fullstops))
```

```python
ocr_list_male = []
male_nyt_ids = []
ocr_list_female = []
female_nyt_ids = []
for i in male_rows.iterrows():
    row = Metadata().query.filter(Metadata.id == int(i[1][0])).one_or_none()
    ocr_list_male.append(row.ocr_transcription)
    male_nyt_ids.append(row.nyt_id)
```

```python
for i in female_rows.iterrows():
    row = Metadata().query.filter(Metadata.id == int(i[1][0])).one_or_none()
    ocr_list_female.append(row.ocr_transcription)
    female_nyt_ids.append(row.nyt_id)
```

```python
ocr_cleaned_male = clean_text(ocr_list_male)
ocr_cleaned_female = clean_text(ocr_list_female)
```

```python
from collections import Counter 
ocr_counters_male = [Counter(i) for i in ocr_cleaned_male]
ocr_counters_female = [Counter(i) for i in ocr_cleaned_female]
ocr_counters_all = []
for i in ocr_counters_male:
    ocr_counters_all.append(i)
for i in ocr_counters_female:
    ocr_counters_all.append(i)
nyt_ids_all = []
for i in male_nyt_ids:
    nyt_ids_all.append(i)
for i in female_nyt_ids:
    nyt_ids_all.append(i)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from application.selective_features import dictionaries_without_features

ocr_counters_no_stops = dictionaries_without_features(ocr_counters_all, fullstops)

v = DictVectorizer()
X = v.fit_transform(ocr_counters_no_stops)
#X = v.fit_transform(ocr_counters_all)
y = TfidfTransformer()
Z = y.fit_transform(X)

scaled_vsm = Z.toarray()
#v.vocabulary_["she"] #Feature number outputed by .vocabulary is the index of the searched word in the array.
#len(ocr_counters_no_stops)
```

```python
try:
    from sklearn.model_selection import train_test_split
except:
    from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression

rows = []
y_as_list = list(range(len(scaled_vsm)))
for i in range(500):
    train_size_int = 435
    test_size_int = 100
    X_train, X_test, y_train, y_test = train_test_split(scaled_vsm, y_as_list, train_size=train_size_int, \
                                                        test_size=test_size_int, random_state=i)
    train_labels = []
    # 0 = male here
    for pos in y_train:
        if pos < 414:
            value = 0
        else:
            value = 1
        train_labels.append(value)
    test_labels = []
    for pos in y_test:
        if pos < 414:
            value = 0
        else:
            value = 1
        test_labels.append(value)
    # Create classifiers
    lr = LogisticRegression()
    lr.fit(X_train, train_labels)
    results = lr.predict(X_test)
    score = lr.score(X_test, test_labels)
    y_train = [str(u) for u in y_train]
    y_test = [str(z) for z in y_test]
    a = ', '.join(y_train)
    b = ", ".join(y_test)
    row = [i, a, b, score, train_size_int, test_size_int, "tfidf_full_text"]
    rows.append(row)
```

```python
cols=["seed", "train_ids", "test_ids", "rate", "train_size", "test_size", "features_used"]
df_results = pd.DataFrame.from_records(rows, columns=cols)
df_results.to_csv("logistic_w_function_words_500_samples_435_train_no_pronouns_or_names.csv")
print(df_results['rate'].describe())
import matplotlib
%matplotlib inline
df_results.hist(column='rate')
words = v.vocabulary_.keys()

labeled_coefs = {}
for i in words:
    try:
        labeled_coefs[i] = lr.coef_[0][v.vocabulary_[i]]
    except: 
        pass
df = pd.DataFrame.from_records(labeled_coefs.items(), columns =["word", "coefficient"])
df = df.sort_values(by='coefficient')
#df.to_csv("example_coefficients.csv")
import numpy as np
#df['odds'] = np.exp(df['coefficient'])
df.to_csv("no_pronouns_or_names_coefficients.csv")
```

```python
import operator
from scipy.stats.stats import pearsonr, spearmanr
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

tfidf = TfidfTransformer()
vec = DictVectorizer()
vect = vec.fit_transform(ocr_counters_all)
adjusted = tfidf.fit_transform(vect)

term_indices = list(vec.vocabulary_.items())
#alphabetical order
term_indices.sort(key=operator.itemgetter(1))

term_list = [i[0] for i in term_indices]
data = adjusted.toarray()

p_tuples = []

male = ["male" for i in male_nyt_ids]
female = ["female" for i in female_nyt_ids]
genders = []
for i in male:
    genders.append(i)
for i in female:
    genders.append(i)

for column in data.T:
   s, c = spearmanr(column, genders)
   f_tuple = (s,c)
   p_tuples.append(f_tuple)


final_tuples = list(zip(term_list, [i[0] for i in p_tuples], [i[1] for i in p_tuples]))
#convert to pandas df
df = pd.DataFrame(final_tuples, columns=["term", "spearman", "spearman_p_value"])
df.sort_values(by='spearman', ascending=False)
```
