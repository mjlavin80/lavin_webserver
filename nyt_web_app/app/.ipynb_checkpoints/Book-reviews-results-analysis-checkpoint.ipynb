```python
%matplotlib inline
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

np.set_printoptions(precision=2)

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Gender Assumed by Book Reviewer')
    plt.xlabel('Machine Predicted Label')
summaries = [df_main_all_summary, df_main_stopwords_summary,df_main_stop_pro_summary,df_main_stop_pro_names_summary,df_main_sentiment_pos_summary,\
             df_main_sentiment_neg_summary,df_main_sentiment_all_summary]
summary_labels = ["all_cm","stopwords_cm","stop_pro_cm","stop_pro_names_cm","sentiment_pos_cm",\
             "sentiment_neg_cm","sentiment_all_cm"]
```

```python
for h,i in enumerate(summaries):
    stopwords_cm = [[i['male_match']['mean'], i['male_mismatch']['mean']], [i['female_mismatch']['mean'], i['female_match']['mean']]]
    stopwords_cm = [[int(i[0]), int(i[1])] for i in stopwords_cm]
    stopwords_cm = np.array(stopwords_cm)

    plt.figure()
    plot_confusion_matrix(stopwords_cm, classes=['male', 'female'], normalize=False, title='')
    plt.tight_layout()
    plt.savefig("figures/"+summary_labels[h]+"_counts.png", dpi=300, transparent=True)
```

```python
def get_coefs(scenario):
    scenario_dict = {
                    "all": {"_min": 0, "_max":1000},
                    "stop_only": {"_min": 1000, "_max": 2000},
                    "stop_pro": {"_min": 2000, "_max": 3000},
                    "stop_pro_names": {"_min": 3000, "_max": 4000},
                    "sentiment_pos": {"_min": 4000, "_max": 5000},
                    "sentiment_neg": {"_min": 5000, "_max": 6000},
                    "sentiment_all": {"_min": 6000, "_max": 7000},
                    }
    #id range
    _min = scenario_dict[scenario]["_min"]
    _max = scenario_dict[scenario]["_max"]
    #fifty random ids 
    _ids_ = c.execute("""SELECT id FROM main WHERE id > ? and id <= ?""", (_min, _max)).fetchall()
    _ids_ = [i[0] for i in _ids_]
    _ids_.sort()
    coefs_ = []
    #query by ids 
    
    base_query = """SELECT id FROM coefficients WHERE main_id = ? LIMIT 1"""
    base = c.execute(base_query, (str(_ids_[0]),)).fetchone()[0]       
    top = c.execute(base_query, (str(_ids_[-1]),)).fetchone()[0]

    #query
    query = """SELECT main_id, feature, score FROM coefficients WHERE id > ? and id < ?"""
    count = 0 
    rows_coef = []
    for u in range(base,top, 1000):
        coef_set = []
        for coef in c.execute(query, (u,u+1001)):

            # if count == 0:
            #    print("First query finished")
            #    count = 1
            coef_set.append(coef)
        rows_coef.extend(coef_set)
    
    return rows_coef
```

```python
def make_coef_lists_gendered(coefs):
    cols_coef = ["main_id","feature","score"]
    df_coef = pd.DataFrame.from_records(coefs, columns=cols_coef)
    #group by feature, compute average score and odds, skips NAN
    df_grouped = df_coef.groupby(['feature',])['score'].mean().to_frame(name = 'feature_averages').reset_index()
    df_grouped = df_grouped.sort_values(by="feature_averages", ascending=False).reset_index(drop=True)
    df_coef_sample_female = df_grouped.iloc[0:50]
    df_coef_sample_male = df_grouped.iloc[-51:-1].reset_index(drop=True)
    return (df_coef_sample_female, df_coef_sample_male)

def get_coef_by_feature(coefs, feature):
    cols_coef = ["main_id","feature","score"]
    df_coef = pd.DataFrame.from_records(coefs, columns=cols_coef)
    
    #query feature
    df_for_feature = df_coef.loc[(df_coef['feature'] == feature)].reset_index(drop=True)
    
    return df_for_feature
```

```python
from nltk.corpus import wordnet as wn

def replacer(term):
    replace = {'n': 'NOUN', 'v':'VERB', 'a':'ADJ', 's':'ADJ', 'r':'ADV'}
    try:
        t = replace[term]
    except: 
        t = term
    return t



pos_coef = []
for z in [df_coefs[3], df_coefs[4], df_coefs[9], df_coefs[10]]:
    pos_options = {}
    for i in list(z['feature']):
        term = nlp(i)
        pos = [q.pos() for q in wn.synsets(i)]
        pos.append(term[0].pos_)
        pos = list(set([replacer(x) for x in pos]))
        pos.sort()
        for e in pos:
            try:
                pos_options[e].append(i)
            except: 
                pos_options[e] = [i,]
    for pos, wordlist in pos_options.items():
        wordlist = list(set(wordlist))
        wordlist.sort()
        pos_options[pos] = ", ".join(wordlist)
    pos_coef.append(pos_options)
        
```

```python
def get_stats(scenario):
    scenario_dict = {
                    "all": {"_min": 0, "_max":1000},
                    "stop_only": {"_min": 1000, "_max":2000},
                    "stop_pro": {"_min": 2000, "_max": 3000},
                    "stop_pro_names": {"_min": 3000, "_max": 4000},
                    "sentiment_pos": {"_min": 4000, "_max": 5000},
                    "sentiment_neg": {"_min": 5000, "_max": 6000},
                    "sentiment_all": {"_min": 6000, "_max": 7000},
                    }
    #id range
    _min = scenario_dict[scenario]["_min"]
    _max = scenario_dict[scenario]["_max"]
    nyt_ids = c.execute("""SELECT nyt_id FROM results WHERE main_id > ? and main_id <= ? GROUP BY nyt_id""", [_min, _max]).fetchall()
    scenario_stats = []
    for i in nyt_ids:
        m_query = """SELECT nyt_id, predicted_gender, Count(predicted_gender), labeled_gender, AVG(probability_male), AVG(probability_female) \
        FROM results WHERE nyt_id = ? and predicted_gender = 0 and main_id > ? and main_id <= ?"""
        f_query = """SELECT nyt_id, predicted_gender, Count(predicted_gender), labeled_gender, AVG(probability_male), AVG(probability_female) \
        FROM results WHERE nyt_id = ? and predicted_gender = 1 and main_id > ? and main_id <= ?"""
        prob_query = """SELECT AVG(probability_male), AVG(probability_female) FROM results WHERE nyt_id = ? \
        and main_id > ? and main_id <= ?"""
        m_result_rows = c.execute(m_query, [i[0], _min, _max]).fetchall()
        f_result_rows = c.execute(f_query, [i[0], _min, _max]).fetchall()
        mean_probabilities = c.execute(prob_query, [i[0], _min, _max]).fetchall()
        mean_probability_male = mean_probabilities[0][0]
        mean_probability_female = mean_probabilities[0][1]
        if m_result_rows[0][3] != None: 
            labeled_gender = m_result_rows[0][3]
        else:
            labeled_gender = f_result_rows[0][3]
        m_result_rows = [n if n else 0 for n in m_result_rows]
        f_result_rows = [n if n else 0 for n in f_result_rows]
        male_percent = m_result_rows[0][2]*1.0/(m_result_rows[0][2]+f_result_rows[0][2])
        female_percent = f_result_rows[0][2]*1.0/(m_result_rows[0][2]+f_result_rows[0][2])
        stats = (i[0], mean_probability_male, mean_probability_female, male_percent, female_percent, labeled_gender)
        scenario_stats.append(stats)
    return scenario_stats
```

```python
#run this on campus
scenarios = ["all", "stop_only","stop_pro","stop_pro_names","sentiment_pos","sentiment_neg","sentiment_all"]
stats_dataframes_dict = {}
for m in scenarios: 
    scenario_stats = get_stats(m)
    print("Finished getting NYT ID stats for %s scenario" % m)
    df_scenario = pd.DataFrame.from_records(scenario_stats, columns=["nyt_id", "mean_probability_male", \
            "mean_probability_female", "male_percent", "female_percent", "labeled_gender"])
    stats_dataframes_dict[m] = df_scenario
```

```python
# top most 10 consistently calculated (archetypal)
# male with the highest female probability and vice versa (outliers)
archetypes_outliers = {}

from scipy.stats.mstats import zscore

for sc in scenarios:
    #split into having labeled_gender 0 and 1
    male_df = stats_dataframes_dict[sc].loc[stats_dataframes_dict[sc]['labeled_gender'] == 0]    
    male_df = male_df.sort_values(by=["male_percent", "mean_probability_male"], ascending=False).reset_index(drop=True)
    #sort each by male_percent (for 0) or female_percent (1) 
    female_df = stats_dataframes_dict[sc].loc[stats_dataframes_dict[sc]['labeled_gender'] == 1]
    female_df = female_df.sort_values(by=["female_percent", "mean_probability_female"], ascending=False).reset_index(drop=True)
    
    #store data for 10 highest and lowest percent
    result_dict = {"archetypes_m": male_df.iloc[0:10], "archetypes_f": female_df.iloc[0:10]}
    result_dict["outliers_m"] = male_df.iloc[-11:-1].reset_index(drop=True)
    result_dict["outliers_f"] = female_df.iloc[-11:-1].reset_index(drop=True)
    
    #consistently ambiguous probabilities ... promiscuous
    
    #absolute value of z-scores
    male_df['m_prob_z'] = [abs(n) for n in zscore(list(abs(male_df['male_percent'] - male_df['female_percent'])))]
    female_df["f_prob_z"] = [abs(n) for n in zscore(list(abs(female_df['female_percent'] - female_df['male_percent'])))]
    #sort 
    male_df = male_df.sort_values(by='m_prob_z', ascending=False).reset_index(drop=True)
    female_df = female_df.sort_values(by='f_prob_z', ascending=False).reset_index(drop=True)
    #ten smallest values
    result_dict['ambiguous_m'] = male_df.iloc[0:10].reset_index(drop=True)
    result_dict['ambiguous_f'] = female_df.iloc[0:10].reset_index(drop=True)
    
    archetypes_outliers[sc] = result_dict
    
```

```python
rows_extreme = []
for i in scenarios:
    for e in ["archetypes_m", "outliers_m", "ambiguous_m", "archetypes_f", "outliers_f", "ambiguous_f"]:
        row = (i, e, archetypes_outliers[i][e]['nyt_id'].loc[0], archetypes_outliers[i][e]['labeled_gender'].loc[0])
        rows_extreme.append(row)
        row2 = (i, e, archetypes_outliers[i][e]['nyt_id'].loc[1], archetypes_outliers[i][e]['labeled_gender'].loc[1])
        rows_extreme.append(row2)
df_extreme = pd.DataFrame.from_records(rows_extreme, columns=['scenario', 'quantifier', 'nyt_id', 'labeled_gender'])
df_extreme.to_csv("figures/extremes.csv")
```
