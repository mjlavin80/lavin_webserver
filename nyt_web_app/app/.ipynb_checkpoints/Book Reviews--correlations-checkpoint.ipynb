```python
#load ocr_counters_all, fullstops from pickles
import pickle
with open('ocr_counters_all.pickle', 'rb') as handle:
    ocr_counters_all = pickle.load(handle)
with open('fullstops.pickle', 'rb') as handle2:
    fullstops = pickle.load(handle2)
with open('nyt_ids_all.pickle', 'rb') as handle3:
    nyt_ids_all = pickle.load(handle3)
with open('fullstops_and_pronouns.pickle', 'rb') as handle4:
    fullstops_and_pronouns = pickle.load(handle4)
with open('fullstops_pronouns_and_names.pickle', 'rb') as handle5:
    fullstops_pronouns_and_names = pickle.load(handle5)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from application.selective_features import dictionaries_without_features

ocr_counters_no_stops = dictionaries_without_features(ocr_counters_all, fullstops)
ocr_counters_no_stops_or_pro = dictionaries_without_features(ocr_counters_all, fullstops_and_pronouns)
ocr_counters_no_stops_pro_names = dictionaries_without_features(ocr_counters_all, fullstops_pronouns_and_names)

v = DictVectorizer()
X = v.fit_transform(ocr_counters_no_stops)
#X = v.fit_transform(ocr_counters_all)
y = TfidfTransformer()
Z = y.fit_transform(X)
scaled_vsm = Z.toarray()

try:
    from sklearn.model_selection import train_test_split
except:
    from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression

rows = []
y_as_list = list(range(len(scaled_vsm)))

train_size_int = 662
test_size_int = 200
X_train, X_test, y_train, y_test = train_test_split(scaled_vsm, y_as_list, train_size=train_size_int, \
                                                        test_size=test_size_int, random_state=0)
train_labels = []
# 0 = male here
for pos in y_train:
    if pos < 703:
        value = 0
    else:
        value = 1
    train_labels.append(value)
test_labels = []
for pos in y_test:
    if pos < 703:
        value = 0
    else:
        value = 1
    test_labels.append(value)
```

```python
import operator
from scipy.stats.stats import pearsonr, spearmanr


term_indices = list(v.vocabulary_.items())
#v.vocabulary_["she"] #Feature number outputed by .vocabulary is the index of the searched word in the array.

#alphabetical order
term_indices.sort(key=operator.itemgetter(1))

term_list = [i[0] for i in term_indices]

data = X_train
text_labels = ["male" if i==0 else "female" for i in train_labels ]

p_tuples = []
for column in data.T:
    s, c = spearmanr(column, text_labels)
    f_tuple = (s,c)
    p_tuples.append(f_tuple)


final_tuples = list(zip(term_list, [i[0] for i in p_tuples], [i[1] for i in p_tuples]))
```

```python
#convert to pandas df
import pandas as pd
df = pd.DataFrame(final_tuples, columns=["term", "spearman", "spearman_p_value"])
df.sort_values(by='spearman', ascending=False)
```
